### Importing libraries and initializing variables

```python
import pandas as pd
from IPython.display import display

global_username = "admin"
global_password = "admin"
df = pd.read_csv(r"HeritageSites.csv")
df = df.drop('Unnamed: 0', axis=1)
```

### 2. Function definitions
#### User Type

```python
# User menu 
def entry():
    print("\n------------------------------")
    print("‚úèÔ∏è Please enter 1 if you are a visitor, else please enter 2 if you are an admin, if you want to quit enter 3: ")
    user_type = int(input())
    if user_type ==1:
        visitor_menu()
    elif user_type == 2:
        admin_info()
    elif user_type == 3:
        exit(1)
        print("You have exited the program")
        print("------------------------------")
    else:
        print("‚ö†Ô∏è Wrong input, try again")
        print("------------------------------")
        entry()
```

#### Admin Functions

```python
# Accepts admin credentials and validates
# If successful, displays Admin Menu
def admin_info():
    entered_username = input("Enter admin username: ")
    entered_password = input("Enter admin password: ")
    print('\n')
    if global_username == entered_username:
        if global_password == entered_password:
            print("You have successfully logged in ü•≥")
            admin_menu()
        else:
            print("Incorrect password ü´®")
            print("------------------------------")
    else:
        print("Incorrect username ü´®")
        print("------------------------------")


# Displays admin menu
def admin_menu():
    print('\n')
    print("#--------- Admin Menu ---------#\n")
    print("1. Add heritage site")
    print("2. Update heritage site")
    print("3. Delete heritage site")
    choice = int(input("Enter the number of your choice from the above menu: "))
    if choice == 1:
        add_site()
    elif choice == 2:
        update_site()
    elif choice == 3:
        delete_site()
        entry()
        

#Admin function to add new heritage site
def add_site():
    add_name = input("Enter new name: ")
    add_State_District = input("Enter the state/district of the new site: ")
    add_year = int(input("Enter the year of the new site: "))
    add_founder = input("Enter founder of the new site: ")
    add_Reason_for_est = input("Enter reason for establishment of the new site: ")
    add_website = input("Enter website: ")
    new_row = {'Name':add_name, 'State_District':add_State_District , 'Year': add_year, 'Founder': add_founder, 'Reason_for_establishment':add_Reason_for_est, 'Website':add_website}
    df.loc[len(df)] = new_row
    df.to_csv("HeritageSites.csv")

    print()
    print("‚úÖ You have successfully added a new heritage site")
    display(df)
    entry()
    
# Admin function to update heritage site information
def update_site():
    site = input("Enter site to be updated: ")
    print("\n", "--------- MENU ---------")
    print('\n', "1. Name", '\n', "2. State_District", '\n', "3. Year", '\n', "4. Founder", '\n', "5. Reason_for_Establishment", '\n', "6. Website", '\n',)
    col_to_modify = int(input("Enter column number to be updated: "))
    new_val = input("Enter new value: ")
    
    dict = {1:"Name", 2:"State_District", 3:"Year", 4:"Founder", 5:"Reason_for_Establishment", 6:"Website"}
    
    for i in range(len(df)):
        if df['Name'].loc[i] == site:
            df[dict[col_to_modify]].loc[i] = new_val
            df.to_csv("HeritageSites.csv")
            
    print()
    print("‚úÖ You have successfully updated the row")
    display(df)
    entry()
    
# Admin function to update heritage site information
def delete_site():
    global df
    display(df)          
    site = input("Enter site to be deleted: ")
    df_deleted = df[df["Name"] != site].reset_index(drop=True)
    df_deleted.to_csv("HeritageSites.csv")
    df = df_deleted
    print()
    print("‚úÖ You have successfully deleted the row")
    display(df_deleted) 
    return df            
```

#### User Functions

```python
# Displays visitor menu
def visitor_menu():
    print('\n')
    print("#--------- Visitor Menu ---------#\n")
    print("1. View all heritage site details")
    print("2. View details for selected heritage site")
    print("3. View all sites for selected state/district")
    choice = int(input("Enter the number of your choice from the above menu: "))
    if choice == 1:
        view_df()
    elif choice == 2:
        view_site()
    elif choice == 3:
        view_details()
        
# User function to display all the heritage sites
def view_df():
    display(df)
    entry()
    
# User function to display information on the heritage site inputted
def view_site():
    site = input("Enter site to be searched: ")
    
    if site not in df["Name"].values:
        print("\n‚ö†Ô∏è Please check entered site name, does not exist")
    else:
        for i in range(len(df)):
            if df["Name"].loc[i] == site:
                display(df.loc[i].to_frame().T)
    entry()
    
# User function to display names of sites present in the inputted state
def view_details():
    val = input("Enter state to be searched: ")
    
    if val not in df["State_District"].values:
        print("\n‚ö†Ô∏è ", val, "does not have a heritage site!")
    else:
        for i in range(len(df)):
            if df['State_District'].loc[i] == val:
                display(df["Name"].loc[i].to_frame().T)
    entry()
  
```
